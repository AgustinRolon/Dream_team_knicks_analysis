## Ahora creare una player_id para todos los jugadores teniendo en cuenta que hay jugadores que se  repiten. A estos hay que asignarle la misma id cada vez.

```python
# Obtén los jugadores únicos
unique_players = df_filtered['Player'].unique()

# Crea un diccionario para mapear nombres de jugadores a IDs
player_id_map = {player: idx + 1 for idx, player in enumerate(unique_players)}

# Asigna IDs a los jugadores en el DataFrame
df_filtered['player_id'] = df_filtered['Player'].map(player_id_map)
```

```python
np.shape(df_filtered)
```

```python
df_filtered.isnull().sum()
```

```python
df_filtered.columns
```

```python
# Agrupamos por 'player_id' y verificamos si cada grupo tiene un solo nombre único en la columna 'Player'
validation = df_filtered.groupby('player_id')['Player'].nunique()

# Filtramos para encontrar cualquier 'player_id' que tenga más de un nombre asociado
invalid_ids = validation[validation > 1]

if invalid_ids.empty:
    print("¡Éxito! Todos los player_id tienen un único nombre de jugador asignado.")
else:
    print("Se encontraron inconsistencias en los siguientes player_id:")
    print(invalid_ids)
```

### Se crea una llave primaria compuesta unificada con la id de jugadores y la temporada en la que jugó.

Esto se decidió para que la tabla de hechos tenga todas ids unicas.

```python
# Crear una nueva columna 'primary_key' que combine 'id' y 'Temp'
df_filtered['primary_key_composed'] = df_filtered['player_id'].astype(str) + "_" + df_filtered['Temp'].astype(str)

# Verificar las primeras filas de la tabla con la nueva columna
print(df_filtered[['player_id', 'Temp', 'primary_key_composed']].head())
```

```python
# Verificar si hay duplicados en la columna 'id'
duplicated_ids = df_filtered[df_filtered['primary_key_composed'].duplicated(keep=False)]

# Mostrar los duplicados encontrados (si existen)
if not duplicated_ids.empty:
    print("Se encontraron los siguientes duplicados en la columna 'primary_key_composed':")
    print(duplicated_ids)
else:
    print("No se encontraron duplicados en la columna 'primary_key_composed'.")

```
