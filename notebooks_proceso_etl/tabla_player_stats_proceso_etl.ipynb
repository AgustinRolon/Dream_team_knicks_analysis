### Observamos que hay jugadores que tenemos datos, pero que todavia no tienen una ID. Procedemos a otorgarsela en forma ascendente (+1) empezando por el ultimo nombrado en nuestra tabla de IDS oficial.

```python
# Identificar las filas donde 'id' es NaN
missing_ids = merged_df['player_id'].isna()

# Generar un id único para estos jugadores, sumando 1 al id más alto existente
# Asignar un id único para los valores NaN comenzando desde el máximo id actual + 1
max_id = merged_df['player_id'].max()
new_ids = np.arange(max_id + 1, max_id + 1 + missing_ids.sum())

# Asignar los nuevos ids a las filas con NaN
merged_df.loc[missing_ids, 'player_id'] = new_ids

# Verificar el resultado
print(merged_df[missing_ids].head())
```
